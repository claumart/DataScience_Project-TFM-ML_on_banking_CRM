# Creación de Clustering

#### KMeans Cluster

Primero definimos el número de clusters para el algoritmo a través del metodo de la silueta

```python
kmeans_silueta = []
for i in range (2, 11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(dataset_transformado)
    labels = kmeans.labels_
    kmeans_silueta.append(silhouette_score(dataset_transformado, labels, metric='euclidean'))
```

```python
print(kmeans_silueta)
print(max(kmeans_silueta))
```

```python
pyplot.plot(range(2, 9), kmeans_silueta[0:7], 'bx-')
pyplot.title('Método de la Silueta para KMeans')
pyplot.xlabel('Número de Clusters')
pyplot.ylabel('Puntaje de la Silueta')
pyplot.show()
```

#### Cluster DBSCAN

```python
#Estos valores de eps y min_samples fueron seleccionados previo a realizar pruebas y determinar que con valores 
#muy bajos de epsilón se generaba una cantidad muy grande de clusters y con valores iguales o mayores a 2.3 el algoritmo se
#quedaba sin recursos para poder realizar el procesamiento de datos.
#Los valores de min_samples se selecionaron en base a que combiandos con los valores aceptables de epsilon, generaban
#una cantidad razonable de clusters (ni muy pocos y muchos)
dbscan_silueta = []
eps_list=[0.5, 0.75, 1, 1.25]
min_sample_list= [1000, 1250, 1500]
dbscan_config = []
for i in eps_list:
    for j in min_sample_list:
        dbscan = DBSCAN(eps=i, min_samples=j).fit(dataset_transformado)
        labels = dbscan.labels_
        dbscan_silueta.append(silhouette_score(dataset_transformado, labels))
        dbscan_config.append(str(i) + '-' + str(j))
```
